```python
import pandas as pd 
import geopandas

df = pd.read_csv("COVID-19 Vaccinations by Residence Current County Health.csv", usecols = ['date_saved','county','county_population','fully_covered'])
query_mask = df['date_saved'].str.contains("2022") 
year = df[query_mask]
year.sort_values(by=['fully_covered'], ascending=False)

df = pd.read_csv("COVID-19 Vaccinations by Residence Current County Health.csv", usecols = ['date_saved','county','county_population','fully_covered'])
ratio = df['fully_covered']/df['county_population'] * 100
ratio = pd.DataFrame(ratio, columns = ['ratio']) 

percent = year.merge(ratio, left_index=True, right_index=True)

percent
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp") # read in the shapefile
neighborhoods.plot() 

vacc_map = neighborhoods.merge(percent, how='left', left_on='hood', right_on='county')
vacc_map.plot(column='ratio', # set the data to be used for coloring
               cmap='OrRd',              # choose a color palette
               edgecolor="white",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "Percent of people vaccinated"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
               )
```

From the data, it looks like Allegheny county has the highest percentage of people vaccinated and is the healthiest neighborhood on the list.
