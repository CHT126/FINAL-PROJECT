## Metric 1

### Introduction
Staying healthy is important for any neighborhood. One best method to stay healthy is get daily vaccinations and boosters.Vaccines have greatly reduced or eliminated many infectious diseases have harmed people. If a majority of the neighborhood is fully vaccinated then its safe to say that infection and viruses would least likely spread overall creating a healthy neighborhood.

### The Metric
The metric I decided on comparing the total amount of people who are full vaccinated (booster and all), to those who are not fully vaccinated. I wanted to get the ratio of those who are fully vaccinated and compare that to the rest of their population. By doing so it will help determine which neighborhood is the healthiest and is least likely to fall ill to a viral outbreak.
### The dataset
I used the NV.csv because it has the data all different vaccination levels of the county population. It counts all the people that are vaccinated and their levels of vaccination. I wanted to take that data and see how many people in each neighborhood was fully vaxed compared to their population

First, I went and imported the vaccine data into my notebook. This data had the vaccinations of everybody in each neighborhood and their levels of vaccination. After that I imported pandas as pd and had it read the vaccine data. I named this df.

```python
#Getting the data set 

import geopandas as pd 
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv("NV.csv")
df
```

After that, I wanted to get the total amount of people vaccinated in each area, so I once agained called df and this time reassigned it to use all of the columns. I created a new value called 'total' and set that equal to 'total_partially_vaccinated' + 'total_fully_vaccinated' + 'total_fully_vaccinated_w_booster'. This way all the columns add up and I get the total amount of people vaccinated in each neighborhood. After that I created a brand new column for total and the output is what you see below.

```python
#gets the total of people vacced

df = pd.read_csv("NV.csv", usecols = ['neighborhood_municipality','total_partially_vaccinated','total_fully_vaccinated','total_fully_vaccinated_w_booster'])
total = df['total_partially_vaccinated']+ df['total_fully_vaccinated'] + df['total_fully_vaccinated_w_booster']
total = pd.DataFrame(total, columns = ['total']) 
total
```

I wanted to merge the new total column with the rest of the data. So I did df.merge to do so and renamed this new graph as 'pop'

```python
# graph with the totals 
pop = df.merge(total, left_index=True, right_index=True)
pop
```

I needed to also get the total ratio of people who were fully vaccinated with the booster shot relative to their population. So I created another new value called 'ratio' and had it set to equal pop['total_fully_vaccinated_w_booster']/pop['total'] * 100. I had to include pop at the begininng to make sure ratio was using the new graph with total in it. In the end I create a percentage of people who were fully vaccinated with booster relative to their respective population. After ratio was done I wanted to create a new column for it outputting what you see below.

```python
# getting percentage of fully vacced with booster compared to rest of the data 

ratio = pop['total_fully_vaccinated_w_booster']/pop['total'] * 100
ratio = pd.DataFrame(ratio, columns = ['ratio']) 
ratio
```

I wanted to merge this new ratio column with the rest of the data set, so I did pop.merge and create the dataset seen below.

```python
#graph percentage 
percent = pop.merge(ratio, left_index=True, right_index=True)
percent
```

So far we just have a big table with numbers all over the place. Why not make a graph so it will be easier understand? I created a bar graph with the neighborhoods as my x and the percentage of people fully vaccinated with booster as my y. In the end I got a nice looking bar as seen below.

```python
#graph of the percentages 
percent.sort_values(by = ['ratio'], ascending = False).head(20).plot.bar(x = 'neighborhood_municipality', y = 'ratio')
```

From the vaccinated data, it looks like Trafford, Regent Square, and Aleppo have the highest percentage of people fully vaccinated.

## Metric 3


### Introduction
The environment of a neighorhood is important to the living quality of people in the neighborhood and the living quality of people the most important factor that determines a good neighborhood. Some neighborhood in large cities like San Francisco and Los Angeles are heavily polluted by people throwing their trash on the ground. This can be caused by the lacking of trash can in the neighborhood. A simple smart trash container can have a huge impact on a city's environment. It can decrease the level of pollution of a neighborhood which can ended up improve people's living quality. 

### The Metric 
The metric will measure the bestness of Pittsburgh neighborhood based on the ratio between the count of the smart trash containers and the population of the neighborhood. If the ratio is high, it means that the people can have access to the smart trash container in the neighborhood and people can have a place to throw out their trash to the trash container instead of throwing it to somewhere else that damages the enviorment. 

### Dataset 
The small_trash_container.csv file contains the data of the location of the city of Pittsburgh's smart trash containers. There are 8 attribute in this dataset: container_id, receptacle_model_id, assignment_date, group_name, address, city, state, zip, neighborhood, dpw_division, council_district, ward, fire_zone, x, y. The useful attribute for this project will be the neighborhood because the metric need to group the count of the containers by neighborhood. The population.csv contains the population of the Pittsburgh neighborhood in different time period from 2010 to 2020. In this project, the useful attributes is the lastest total population which is 2020_total_population.

### Code

Here we call the read_csv function on the small_trash_containers.csv to read the dataset. The unique function shows all the unique neighborhoods in the dataset. 

```python
#Here we import pandas
import pandas as pd
#Call read_csv function to read the container dataset
data = pd.read_csv("smart_trash_containers.csv", usecols = ['neighborhood'])
#All the different neighborhood in the dataset
data['neighborhood'].unique()
```

Then we call the read_csv function on the population.csv. To get the ratio we need the population of the neighborhood. However, some neighborhood in the smart_trash_containers.csv is not in the population.csv. So we need to find the neighborhood that is not in both datasets. The for loop will create a new entry in the dict d if there is currently no entry with the neighborhood name. It will increase the value of the entry if the neighborhood is in the population.csv and in smart_trash_containers.csv. Once the for loop finished running. All the neighborhoods with value 1 are the neighborhood that does not exist in both datasets. We need to remove those entries.

```python
#read population dataset
p = pd.read_csv("population.csv")
d = dict()
#Some data are missing in the population dataset. We need to find all the missing data. In this case all the neighborhood with value 1.
for index, line in data.iterrows():
    if line['neighborhood'] in d and line['neighborhood'] in p['Neighborhood'].unique():
        d[line['neighborhood']] += 1
    else:
        d[line['neighborhood']] = 1
print(d)
```

Here we put all the neighborhoods with value 1 into a list and then delete those neighborhood in dict d.

```python
#Get rid of the value 1 neighborhood
del_list = []
for name in d:
    if d[name] == 1:
        del_list.append(name)
        
d = dict([(key, val) for key, val in 
           d.items() if key not in del_list])
print(d)
```

Now we can divide the number of smart trash containers in each neighborhood with its population. The for loop divide every entry's value in dict d and divided by the neighborhood population.

```python
#Divide the number of smart trash containers by the population to get the ratio
for index, line in p.iterrows():
    if line['Neighborhood'] in d:
        d[line['Neighborhood']] = d[line['Neighborhood']] / line['2020_Total_Population']
print(d)
```

### Result

To make the result easier to read, we call DataFrame function from pandas to make the result display as a table.

```python
#Call DataFrame function to return the result
res = pd.DataFrame(list(d.items()), columns = ['Neighborhood', 'SmartContainerRatio']) 
res
```

Now we sort he values and plot the result to a bar graph.

```python
#plot the result
res.sort_values(by = ['SmartContainerRatio'], ascending = False).plot.bar(x = 'Neighborhood', y = 'SmartContainerRatio')
```

There are a lot of neighborhoods on the x axis. Let's look at the top 20 results by using .head(20).

```python
res.sort_values(by = ['SmartContainerRatio'], ascending = False).head(20).plot.bar(x = 'Neighborhood', y = 'SmartContainerRatio')
```

Now we plot the result to a map. We can use geopandas to do the work. We first create neighborhoods using read_files function. Then we use merge function to merge our results with the neighborhoods. Lets look at the head to confirm it merged correctly.

```python
import geopandas
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
# do the merge
steps_map = neighborhoods.merge(res, how='left', left_on='hood', right_on='Neighborhood')
# look at the head to confirm it merged correctly
steps_map[['hood','Neighborhood', 'SmartContainerRatio']].head()
```

Then we plot the map using color green.

```python
steps_map.plot(column='SmartContainerRatio', # set the data to be used for coloring
               cmap='Greens',              # choose a color palette
               edgecolor="black",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "SmartContainerRatio"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
               )
```

According to the result plot, North shore has the highest ratio between the neighborhood population and smart trash containers.
